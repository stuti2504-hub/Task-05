```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("customer_sales.csv")

# Total Sales
total_sales = df["Total_Sales"].sum()
print(f"Total Sales: â‚¹{total_sales:,}")

# Top 5 Customers
top_customers = df.groupby("Customer_Name")["Total_Sales"].sum().sort_values(ascending=False).head(5)
print("\nTop Customers:\n", top_customers)

# Product Sales Analysis
product_sales = df.groupby("Product")["Total_Sales"].sum().sort_values(ascending=False)

# Region-wise Sales
region_sales = df.groupby("Region")["Total_Sales"].sum()

# Monthly Trend
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.to_period("M")
monthly_sales = df.groupby("Month")["Total_Sales"].sum()

# ---------- Visualization ----------

# 1. Top Customers Bar Chart
top_customers.plot(kind='bar', color='skyblue', title="Top 5 Customers by Sales")
plt.ylabel("Total Sales (â‚¹)")
plt.show()

# 2. Product Sales Pie Chart
product_sales.plot(kind='pie', autopct='%1.1f%%', title="Product Sales Distribution")
plt.ylabel("")
plt.show()

# 3. Region-wise Sales Bar Chart
region_sales.plot(kind='bar', color='lightgreen', title="Sales by Region")
plt.ylabel("Total Sales (â‚¹)")
plt.show()

# 4. Monthly Sales Line Chart
monthly_sales.plot(kind='line', marker='o', title="Monthly Sales Trend", color='orange')
plt.ylabel("Total Sales (â‚¹)")
plt.show()
```
